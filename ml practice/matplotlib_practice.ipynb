```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.log(x)
```

```python
plt.plot(x,y)
plt.title('Log of the values')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim(0,6)
plt.ylim(0,2)
plt.savefig('firstplot.png')
```

```python
fig = plt.figure(figsize=(10,10))
axes = fig.add_axes([0.5,0.5,0.5,0.5])
axes.plot(x,y)
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig = plt.figure() # 640x480
# Large
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#Small
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x,y)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')

plt.show()
```

```python
fig = plt.figure(figsize=(2,2),dpi=200) # 640x480
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
# fig.savefig('newfig.png',bbox_inches='tight')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
# fig.tight_layout()
fig.subplots_adjust(wspace=1)
fig.suptitle('Hello',fontsize=16)
# plt.show()
fig.savefig('newfig.png',bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X Vs X',color='#e6c017',lw=10,ls='--')
# lines = ax.plot(x,x**4, label= 'X Vs X^2',color='purple')
# lines[0].set_dashes([5,2,10,2])
ax.legend(loc=(0.5,0.5))
ax.plot(x,x**4, label= 'X Vs X^2',color='purple',marker='o',markersize=20,markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange')
```
