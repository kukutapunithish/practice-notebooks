```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Excercises/08-Linear-Regression-Models/Advertising.csv')
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_net_model = ElasticNet()
```

```python
param_grid = {
    'alpha': [0.1,1,5,10,50,100],
    'l1_ratio': [.1,.5,.7,.95,.99,1]
}
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns = ['sales'])
Y = df['sales']
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=47)
# model = LinearRegression()
# model.fit(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,param_grid=param_grid,scoring='neg_root_mean_squared_error',cv=5,verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
met = pd.DataFrame(grid_model.cv_results_)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error

def metrics(actual,predictions):
    mae = mean_absolute_error(actual,predictions)
    print('mae ',mae)
    mse = mean_squared_error(actual,predictions)
    print('mse ',mse)
    rmse = root_mean_squared_error(actual,predictions)
    print('rmse ',rmse)
    return mae,mse,rmse
```

```python
mae,mse,rmse = metrics(y_test,grid_model.predict(X_test))
```
