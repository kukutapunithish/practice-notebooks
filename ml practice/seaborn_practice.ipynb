```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Excercises/05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
sns.scatterplot(x='salary',y='sales',hue='level of education',style='level of education',data=df)
```

```python
sns.scatterplot(x='salary',y='sales',hue='salary',data=df)
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='whitegrid')
sns.displot(data=df,x='salary',bins=20,kde=True)
```

```python
sns.histplot(data=df,x='salary')
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,errorbar='sd')
```

```python
temp = df.groupby('level of education')['salary'].std()
```

```python
plt.pie(temp.values,labels=temp.index.values)
plt.show()
```

```python
df = pd.read_csv('../Excercises/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.boxplot(data=df,y='math score')
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,y='math score',x='parental level of education',hue='test preparation course')
```

```python
sns.violinplot(data=df,x='math score',y='parental level of education')
```

```python
sns.swarmplot(data=df,x='math score',size=3)
```

```python
sns.boxenplot(x='math score',y='test preparation course',data=df)
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',fill=True,hue='gender')
# 
```

```python
sns.pairplot(data=df,hue='gender',corner=True)
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='race/ethnicity')
```

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
```
