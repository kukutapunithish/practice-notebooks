```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Excercises/05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='salary',y='sales',hue='level of education',style='level of education',data=df)
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='whitegrid')
sns.displot(data=df,x='salary',bins=20,kde=True)
```

```python
sns.histplot(data=df,x='salary')
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,x='level of education',y='salary',estimator=np.std)
```

```python
df.groupby('level of education')['salary'].std()
```
